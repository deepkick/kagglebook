```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
```

```python
##################################
# [1] データの確認
# https://www.kaggle.com/c/titanic/data から取得
##################################
data_df = pd.read_csv("titanic_csv/train.csv")
data_df
```

```python
##################################
# [2] データの基本的な統計量等を確認
##################################
data_df.describe()
```

```python
##################################
# [3] データの前処理
##################################
# [3.1] 各特徴のデータ型を確認
##################################

data_df.info()
```

```python
##################################
# [3.2] 欠損値の数を確認
##################################

data_df.isnull().sum()
```

```python
##################################
# [3.3] 方針の決定と処理
# Age ,Cabin, Embarkedが、行数と整合せずに欠損している。
# 方針：Cabinは欠損件数が多いので列を削除。Age, Embarkedについては、欠損している行を削除。
##################################

data_df = data_df.drop(["Cabin"], axis = 1)
data_df = data_df.dropna(subset = ["Age", "Embarked"])
data_df
```

```python
##################################
# [3.４] 予測に無関係な列を消去
##################################

data_df = data_df.drop(["PassengerId", "Name", "Ticket"], axis = 1)
data_df.dtypes
```

```python
##################################
# [3.5] 予カテゴリ変数をダミー変数に変換する
##################################

cat_col = ['Sex', 'Pclass', 'Embarked']
data_df = pd.get_dummies(data_df, columns = cat_col)
data_df.dtypes
```

```python
data_df
```

```python
##################################
# [4] 訓練データとテストデータの分割
# データを8:2 に分割し、それぞれ訓練データ、テストデータとする
# 訓練データで学習をし、テストデータで分析手法の評価をする
##################################


train_df, test_df = train_test_split(data_df, test_size = 0.2, random_state = 0)
print('訓練データ数: {}, テストデータ数: {}' .format(len(train_df), len(test_df)))
```

```python
train_df
```

```python
test_df
```

```python
##################################
# [5] データの標準化
# データを標準化（各列の平均を0, 分散を1）する。
# 標準化のための変換方法は、訓練データでの平均・分散から定める
# テストデータについては、一般に訓練データの平均・分散で代用した標準化を行う。
##################################



# numpyの配列として値を取り出す
X_train = train_df.iloc[:, 1:].values
Y_train = train_df.iloc[:,0].values
X_test = test_df.iloc[:, 1:].values
Y_test = test_df.iloc[:,0].values
print("X_train", X_train)
print("Y_train", Y_train)
print("X_test", X_test)
print("Y_test", Y_test)

# fit_transform は与えられたデータの平均・分散の計算と標準化を実行する
# transformではfit_transformで計算された平均・分散を用いた標準化を与えられたデータに対して実行する

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
##################################
# [6] ロジスティック回帰による分類
# 正則化係数は5-交差検証で決定する
# GridSearchCVでは、fitで交差検証が行われ、選ばれたモデルがデフォルトに設定される
##################################



params = {"C": np.logspace(0, 4, 5)}
logreg_cv = GridSearchCV(LogisticRegression(), cv = 5, param_grid = params)
logreg_cv.fit(X_train, Y_train)
print('訓練データでの分類精度：{0:.2%}'.format(logreg_cv.score(X_train, Y_train)))
print('テストデータでの分類精度：{0:.2%}'.format(logreg_cv.score(X_test, Y_test)))
```

```python
##################################
# [7] 混同行列の描画
# Matplotlibライブラリは豊富なグラフ描画機能を提供。
# 分類結果の混同行列も描画
##################################

cm = confusion_matrix(Y_test, logreg_cv.predict(X_test), labels = [0, 1])

plt.xticks( [0, 1, 2], [0, 1])
plt.yticks( [0, 1, 2], [0, 1])
plt.title('Confusion matrix')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')

plt.imshow(cm, cmap = plt.cm.inferno)
plt.colorbar()
```
